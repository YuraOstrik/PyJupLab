```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import kagglehub
from ipywidgets import widgets, interactive_output, HBox, VBox
from IPython.display import display

dataset_path = kagglehub.dataset_download("nalisha/job-salary-prediction-dataset")
csv_file = os.path.join(dataset_path, "job_salary_prediction_dataset.csv")
dataframe = pd.read_csv(csv_file)
chart_type = widgets.Dropdown(options=["Line", "Bar", "Scatter"], value="Bar", description="Plot:")
x_axis = widgets.Dropdown(options=dataframe.columns, description="X:")
y_axis = widgets.Dropdown(options=dataframe.select_dtypes(include="number").columns, description="Y:")
use_grouping = widgets.Checkbox(value=True, description="Group")
aggregation = widgets.Dropdown(options=["mean", "sum", "min", "max", "count"], value="mean", description="Func:")

enable_sort = widgets.Checkbox(value=True, description="Sort")
sort_order = widgets.Dropdown(
    options=["Ascending", "Descending"],
    value="Ascending",
    description="Order:"
)
range_slider = widgets.IntRangeSlider(
    value=[0, 50],
    min=0,
    max=len(dataframe),
    step=1,
    description="Range:",
    continuous_update=False
)

save_button = widgets.Button(description="Save Plot", button_style='success', icon='save')
top_ui = HBox([chart_type, x_axis, y_axis])
middle_ui = HBox([use_grouping, aggregation, enable_sort, sort_order])
bottom_ui = HBox([range_slider, save_button])
main_ui = VBox([top_ui, middle_ui, bottom_ui])
current_figure = None

def update_chart(chart_type, x_axis, y_axis, use_grouping, aggregation, enable_sort, sort_order, range_slider):
    global current_figure
    plt.close('all')    
    fig, ax = plt.subplots(figsize=(12, 6))
    current_figure = fig

    if chart_type == "Line":
        draw_func = ax.plot
    elif chart_type == "Bar":
        draw_func = ax.bar
    elif chart_type == "Scatter":
        draw_func = ax.scatter
    else:
        return

    ascending = True if sort_order == "Ascending" else False
    if use_grouping:
        data = dataframe.groupby(x_axis)[y_axis].agg(aggregation)
        if enable_sort:
            data = data.sort_values(ascending=ascending)
        data = data.iloc[range_slider[0]:range_slider[1]]
        x_vals = data.index.astype(str)
        y_vals = data.values
        title = f"{chart_type}: {aggregation.title()} {y_axis} by {x_axis}"
    
    else:
        data = dataframe.copy()
        if enable_sort:
            data = data.sort_values(by=y_axis, ascending=ascending)
        data = data.iloc[range_slider[0]:range_slider[1]]
        x_vals = data[x_axis].astype(str)
        y_vals = data[y_axis]
        title = f"{chart_type}: {y_axis} by {x_axis}"
    draw_func(x_vals, y_vals)
    ax.set_title(title, fontsize=14, fontweight='bold')
    ax.set_xlabel(x_axis)
    ax.set_ylabel(y_axis)
    plt.xticks(rotation=45, ha='right')
    ax.grid(axis='y', linestyle='--', alpha=0.7)
    record_count = len(data)
    ax.text(
        0.01, 0.95,
        f"Records: {record_count}",
        transform=ax.transAxes,
        fontsize=10,
        verticalalignment='top',
        bbox=dict(boxstyle="round", alpha=0.3)
    )
    plt.tight_layout()
    plt.show()
def save_chart(btn):
    if current_figure:
        current_figure.savefig("saved_plot.png", dpi=300, bbox_inches='tight')
        btn.description = "Saved!"
        btn.button_style = 'info'
save_button.on_click(save_chart)
output = interactive_output(
    update_chart,
    {
        "chart_type": chart_type,
        "x_axis": x_axis,
        "y_axis": y_axis,
        "use_grouping": use_grouping,
        "aggregation": aggregation,
        "enable_sort": enable_sort,
        "sort_order": sort_order,
        "range_slider": range_slider
    }
)

display(main_ui, output)
```
